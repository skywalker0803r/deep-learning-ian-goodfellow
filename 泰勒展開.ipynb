```
import numpy as np
from mpmath import diff,sin
import matplotlib.pyplot as plt
import math
```

```
def f(x):
  return sin(x)
```

```
def factorial(n):
    factor = 1
    for i in range(1,n+1):
        factor *= i
    return float(factor)

# check
assert factorial(5) == 5*4*3*2*1
```

```
def Taylor(f,x,N):
  a = 0
  res = f(a) + diff(f,x=a,n=1)*(x-a) # f(x) = f(a) + f'(a)(x-a)
  for n in range(2,N+1):
    term1 = 1.0/factorial(n)
    term2 = diff(f,x=a,n=n)
    term3 = (x-a)**n
    res +=  term1*term2*term3
  return res
```

```
x = np.arange(-10,10,0.01)
plt.figure(figsize=(20,20))
for n in range(2,2+16):
  plt.subplot(4,4,n-1)
  y1 = [f(xi) for xi in x]  
  plt.plot(y1,label='origin')
  y = [Taylor(f,xi,n) for xi in x]
  plt.plot(y,label='Taylor {}'.format(n))
  plt.ylim([-1.5,1.5])
  plt.legend()
plt.show()
```
