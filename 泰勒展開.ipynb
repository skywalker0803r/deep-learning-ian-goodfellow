```
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
def f(x):
  return math.e**x
```

```
def factorial(n):
    factor = 1
    for i in range(1,n+1):
        factor *= i
    return float(factor)
assert factorial(5) == 5*4*3*2*1
```

```
def Taylor(f,x,N):
  
  e = 1e-8
  a = 0
  
  # 計算一階微分
  dy = f(a+e) - f(a)
  dx = e
  d = dy / dx # f'(a)
  total = f(a) + d*(x-a) # f(x) = f(a) + f'(a)*(x-a)
  
  for n in range(2,N+1):
    
    # update total 
    term1 = 1.0 / factorial(n)
    term2 = ((d+e)-(d)) / (e)
    term3 = (x-a)**n
    total +=  term1*term2*term3
    
    # update d
    d = term2
  
  return total
```

```
x = np.arange(-10,10,0.01)
plt.figure(figsize=(10,10))
for n in range(2,2+9):
  plt.subplot(3,3,n-1)
  y1 = [f(xi) for xi in x]  
  plt.plot(y1,label='origin')
  y = [Taylor(f,xi,n) for xi in x]
  plt.plot(y,label='Taylor {}'.format(n))
  plt.ylim([-1000,1000])
  plt.legend()
plt.show()
```
