```python
import numpy as np
import matplotlib.pyplot as plt
def game(init_value,total_money,runs):
  下注金額 = init_value #初始下注金額設定
  history = [] #記錄過程
  for i in range(runs): #開始遊戲
    if np.random.uniform(0,1) < 0.5:
      total_money += 下注金額
      下注金額 = init_value
    else:
      total_money -= 下注金額
      if total_money <=0 :
        info = '爆倉'
        break
      下注金額 *= 2 #失敗後要倍增下注金額
    history.append(total_money)
    info = '成功'
  return history,info

info_list = []
n = 100
finally_money = []
all_history = []
for _ in range(n):
  history,info = game(0.1,30,50)
  if history[-1] > 0:
    finally_money.append(history[-1])
    all_history.extend(history)
  info_list.append(info)
  plt.plot(history)
print(f'玩{n}回合成功下庄率:',info_list.count('成功')/len(info_list))
print('平均最終取回資金',np.mean(finally_money))
print('最大最終取回資金',np.max(finally_money))
print('最大中途取回資金',np.max(all_history))
```
